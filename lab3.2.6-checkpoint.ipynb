# 3.2.6 Исследование гальванометра

## 1)Определение динамической постоянной

I = U_0*(R_1/R_2)*1/(R+R_0)

I=x/2a*C_I

U_0 = 1.32 +- 0.01 B

R_1/R_2=1/2000

R_2 = 10 kOm

R_0 = 500 Om

```python
U_0 = 1.32

ratio=1/2000

R_2 = 10000

R_0 = 500.0 

dR = 1.0 

R11 = np.array([8.9, 9.0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60]) #kOm
x11 = np.array([24.7, 24.5, 22.0, 14.8, 11.3, 9.0, 7.6, 6.5, 5.8, 5.2, 4.7, 4.0])#cm
I11 = np.array([])#A
sigma_I11 = np.array([])

for i in range(len(R11)):
    I11 = np.append(I11, U_0*(ratio)*1/(R11[i]*1000+R_0)*1000000000)
    sigma_I11 = np.append(sigma_I11, I11[i]*np.sqrt(dR**2*(1/(R11[i]*1000)**2+1/R_0**2)+0.01**2/U_0**2))
    print(R11[i])
print(I11)
print(sigma_I11)
```

```python
plt.figure(figsize=(20,14))
plt.xlabel(u'$x,$ см', fontsize=18)
plt.ylabel(u'$I,$ нА', fontsize=18)
plt.xlim([0, 25])
plt.ylim([0, 80])
plt.title(u'График зависимости $I$ от $x$', fontsize=20)
plt.errorbar(x11, I11, xerr = 0.1, yerr = sigma_I11, fmt='none', barsabove=True)
plt.plot(x11, I11, 'o', label='_nolegend_',markersize=6)

def f0(x, b0, a0):
    return x*b0+a0
def f1(x, b0):
    return x*b0

Iopt, Icov = curve_fit(f0,xdata = x11, ydata = I11, sigma = sigma_I11)
x_fit = np.append(x11, 0)
plt.plot(x_fit, f0(x_fit, *Iopt))
print(Iopt)
print(np.sqrt(Icov[0,0]))
print(np.sqrt(Icov[1,1]))
plt.grid()
#plt.savefig("3.2.6.1.png")
```

```python
R22 = np.array([28.5, 42.0, 55.0, 70, 80, 90])
ratio22 = np.array([1/1000, 1/500, 1/500,1/300,1/300,1/200])
x22max = np.array([14.0, 18.6,14.2,18.4,16, 21.2])
x22next =np.array([2, 4.8,4.7,7.3,7.1,9.8])

delta22 = np.array([])
_delta22 = np.array([])
sumR22 = np.array([])

sgm_d22 = np.array([])
sgm_1_d22 = np.array([])

for i in range(len(R22)):
    delta22 = np.append(delta22, np.log(x22max[i]/x22next[i]))
    _delta22 = np.append(_delta22, 1/delta22[i]**2)
    sumR22 = np.append(sumR22, (R22[i]+(R_0/1000))**2)
    
    sgm_d22 = np.append(sgm_d22, 0.1*(1/x22max[i]+1/x22next[i]))
    sgm_1_d22 = np.append(sgm_1_d22, _delta22[i]*2*sgm_d22[i]/delta22[i])
print(delta22)
print(_delta22)
print(sumR22)
print(sgm_d22)
print(sgm_1_d22)
```

```python
plt.figure(figsize=(18,10))
plt.xlabel(u'$(R+R_0)^2,$ кОм$^2$', fontsize=12)
plt.ylabel(u'$1/\delta^2$', fontsize=12)
plt.xlim([0, sumR22[len(R22)-1]+200])
#plt.ylim([0, 80])
plt.title(u'График зависимости $1/\delta^2$ от $(R+R_0)^2$', fontsize=15)
plt.errorbar(sumR22, _delta22, yerr = sgm_1_d22, fmt='none', barsabove=True)
plt.plot(sumR22, _delta22, 'o', label='_nolegend_',markersize=6)

dopt, dcov = curve_fit(f0, xdata = sumR22, ydata = _delta22, sigma = sgm_1_d22)
x_fit = np.append(sumR22, 0)
plt.plot(x_fit, f0(x_fit, *dopt))
print(dopt)
print(np.sqrt(dcov[0,0]))
print(np.sqrt(dcov[1,1]))
plt.grid()
#plt.savefig("3.2.6.2.png")
```

```python
R33 = np.array([50,40,30,20,10,5,3,4,7.5,9,12,15,18,33,37,45,35,13,6,16.5])
l33 = np.array([21.2, 20.1,18.7,16.4,12.8,8.7,6.3,7.3,10.8,12.7,13.6,14.7,15.8,18.8,19.6,20.6,19.4,14.3,9.4,15.5])

sumR33 = np.array([])

for i in range(len(R33)):
    sumR33 = np.append(sumR33, 1/(R33[i]+(R_0/1000)))
    
plt.figure(figsize=(18,10))
plt.xlabel(u'$(R+R_0)^{-1},$ кОм$^{-1}$', fontsize=12)
plt.ylabel(u'$l_{max}$, см', fontsize=12)
#plt.xlim([0, sumR33[len(R33)-1]])
#plt.ylim([0, 80])
plt.title(u'График зависимости $l_{max}$ от $(R+R_0)^{-1}$', fontsize=15)
plt.errorbar(sumR33, l33, yerr = 0.1, fmt='none', barsabove=True)
plt.plot(sumR33, l33, 'o', label='_nolegend_',markersize=6)

x_int = np.linspace(sumR33[0]*(1-0.21), sumR33[6]*(1+0.01), 600)
#f = interpolate.interp1d(sumR33, l33, kind='cubic')

#plt.plot(x_int, f(x_int))

def Fexp(x, a0, a1, a2, a3):
    return (a0*np.exp(a1*x)*1/np.sqrt((a2+a3*x**2)))

eopt, ecov = curve_fit(Fexp, xdata = sumR33, ydata = l33)
plt.plot(x_int, Fexp(x_int, *eopt))

#print(dopt)
#print(np.sqrt(dcov[0,0]))
#print(np.sqrt(dcov[1,1]))
plt.grid()
plt.savefig("3.2.6.3.png")

l_max = 25.2 # cм
l_cr = l_max * (math.e**(0.12/4-1))
#print(l_cr)
# по графику определили
for i in x_int:
    if (Fexp(i, *eopt) < (l_cr + 0.01)) and (Fexp(i, *eopt) > (l_cr - 0.01)):
        print(i)
```
