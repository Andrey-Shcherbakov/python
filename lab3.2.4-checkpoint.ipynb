```python
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.interpolate import interp1d
```

```python
C = [0.02, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.08, 0.15, 0.06, 0.90, 0.75]
x_0 = [48,48,47,48,48,48, 48, 48, 48, 47, 48, 48]
x =[40,44,37,45,45,45,41,36,46,40,43,44,40]
n = [25,13,8,8,7,6,5,15,11,16,5,5]
```

```python
T_0 = 0.01
L = 144.02*0.001
sigma_x = 0.5
sigma_L=0.01*0.001
sigma_C = 0.0001
T_exp = []
T_th = []
sigma_T_exp=[]
sigma_T_th = []
for i in range(len(C)):
    T_exp.append(T_0*x[i]/(x_0[i]*n[i])*1000000)
    T_th.append(2*np.pi*np.sqrt(L*C[i]*0.000001)*1000000)
    sigma_T_exp.append(T_exp[i]*np.sqrt((sigma_x**2)*(1/x[i]**2+1/x_0[1]**2)))
    sigma_T_th.append(T_th[i]*np.sqrt(1/2*(sigma_L**2/L**2+sigma_C**2/C[1]**2)))
sigma_T_th
```

```python
plt.figure(figsize=(14,8))
plt.xlabel(u'$T_{th}, \mu c$')
plt.ylabel(u'$T_{exp}, \mu c$')
plt.xlim([300, 2100])
plt.ylim([300, 2100])
plt.title(u'График зависимости $T_{exp}$ от $T_{th}$')
plt.errorbar(T_th, T_exp, xerr = sigma_T_th, yerr = sigma_T_exp, fmt='none', barsabove=True)
plt.plot(T_th, T_exp, 'o', label='_nolegend_',markersize=4)

from scipy.optimize import curve_fit
def f0(x, b0, a0):
    return x*b0+a0
def f1(x, b0):
    return x*b0
Topt, Tcov = curve_fit(f0,xdata = T_th, ydata = T_exp, sigma = sigma_T_exp)
plt.plot(T_th, f0(np.array(T_th), *Topt))
print(Topt)
print(np.sqrt(Tcov[0,0]))
print(np.sqrt(Tcov[1,1]))
plt.grid()
plt.savefig("3.2.4.1.png")
```
