## 1) Градуировка электромагнита

```python
I_grad = np.array([0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 1.00, 1.20, 1.49, 1.30, 1.40]) #mA
Fi_grad = np.array([0.45, 0.7, 1.4, 2.05, 2.65, 3.35, 3.95, 4.50, 5.20, 5.80, 6.20, 7.05, 7.70, 7.35, 7.55]) # mVb
SN = 72*(10**(-4)) #м^2*вит

plt.figure(figsize=(20,14))
plt.xlabel(u'$I_M,$ мA', fontsize=18)
plt.ylabel(u'$B,$ Тл', fontsize=18)
#plt.xlim([0, 25])
#plt.ylim([0, 80])
plt.title(u'График зависимости $B$ от $I_M$', fontsize=20)
#plt.errorbar(I_grad, Fi_grad, fmt='none', barsabove=True)
plt.plot(I_grad, Fi_grad*0.001/SN, 'o', label='_nolegend_',markersize=6)

def f_grad0(x, b0, a0, c0, d0):
    return (x**3)*b0+(x**2)*a0+x*c0+d0
def f0(x, b0, a0):
    return x*b0+a0

gopt, gcov = curve_fit(f_grad0, xdata = I_grad, ydata = Fi_grad*0.001/SN)#, sigma = sigma_I11)
#I_grad = np.append(I_grad, 0)
x_int = np.linspace(I_grad[0]*(1-0.1), I_grad[14]*(1+0.1), 600)
plt.plot(x_int, f_grad0(x_int, *gopt), 'g')

gkopt, gkcov = curve_fit(f0, xdata = I_grad[:10], ydata = Fi_grad[:10]*0.001/SN)
plt.plot(x_int, f0(x_int, *gkopt), 'r--')

print(gopt)
gerr = np.sqrt(np.diag(gcov))
print(gerr)

print(gkopt)

plt.grid()
#plt.savefig("3.3.4.1.png")
```
