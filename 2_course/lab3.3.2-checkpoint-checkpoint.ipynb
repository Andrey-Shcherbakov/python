```python
xdata = []
x_err =[]
y_err = []
y = I_1
for i in range(len(V)):                              
    xdata.append(V[i]**1.5)
    x_err.append(V[i]*0.005+0.02)
    y_err.append(y[i]*0.005+0.05)

from scipy.optimize import curve_fit
def f0(x, b0, a0):
    return x*b0+a0

plt.figure(figsize=(14,8))
#plt.ylim([0, 500])
#plt.xlim([0, 31])
plt.xlabel(u'$V^{3/2},$ В$^{3/2}$')
plt.ylabel(u'$I, \mu$ A')
plt.title(u'График зависимости $I$ от $V^{3/2}$ на всём диапазоне, ток накала 1.60 А')
plt.errorbar(xdata, y, xerr=x_err, yerr = y_err, fmt='none', barsabove=True)
plt.plot(xdata, y, 'o', label='_nolegend_',markersize=3)


p1opt, p1cov = curve_fit(f0,xdata[16:], ydata = y[16:], sigma = y_err[16:])
plt.plot(xdata[16:], f0(np.array(xdata[16:]), *p1opt))

print(p1opt)
print(np.sqrt(p1cov[0,0]))
print(np.sqrt(p1cov[1,1]))

plt.grid()
#plt.savefig("3.3.2_1.60.png")
```

```python
f1 = interp1d(xdata[0:17], y[0:17], kind='nearest')

plt.xlabel(u'$V^{3/2},$ В$^{3/2}$')
plt.ylabel(u'$I, \mu$ A')
plt.title(u'График зависимости $I$ от $V^{3/2}$ в начале, ток накала 1.30 А')

plt.plot(xdata[0:17], f1(xdata[0:17]), '-', xdata[0:17], y[0:17], 'o')
plt.grid()
plt.savefig("3.3.2_1.30_2.png")
```
